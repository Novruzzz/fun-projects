# First Methodology: MFCCs Visualisation

```python
# RAVDESS; Female; Angry
path = '/content/Actor_08/03-01-05-02-01-01-08.wav'
data, sample_rate = librosa.load(path, sr=44100, offset=0.5, duration=2.5, res_type='kaiser_fast')
mfccs = librosa.feature.mfcc(y=data, sr=sample_rate, n_mfcc=13)

# Acoustic wave
plt.figure(figsize=(20, 15))
plt.subplot(3,1,1)
librosa.display.waveshow(data, sr=sample_rate)
plt.title('Audio sampled at 44100 Hz')

# MFCC
plt.figure(figsize=(20, 15))
plt.subplot(3,1,1)
librosa.display.specshow(mfccs, x_axis='time')
plt.ylabel('MFCCs')
plt.colorbar()

ipd.Audio(path)
```

# Second Methodology: MFCCs Statistics

```python
# RAVDESS; Female; Angry
path = '/content/Actor_08/03-01-05-02-01-01-08.wav'
data, sample_rate = librosa.load(path, res_type='kaiser_fast',duration=2.5,sr=22050*2,offset=0.5)
female = np.mean(librosa.feature.mfcc(y=data, sr=sample_rate, n_mfcc=13), axis=0)
print(len(female))

# RAVDESS; Male; Angry
path = '/content/Actor_09/03-01-05-01-01-01-09.wav'
data, sample_rate = librosa.load(path, res_type='kaiser_fast',duration=2.5,sr=22050*2,offset=0.5)
male = np.mean(librosa.feature.mfcc(y=data, sr=sample_rate, n_mfcc=13), axis=0)
print(len(male))

# Plotting the mean MFCCs with respect to the number of bands
plt.figure(figsize=(20, 15))
plt.subplot(3,1,1)
plt.plot(female, label='female')
plt.plot(male, label='male')
plt.legend()
```
