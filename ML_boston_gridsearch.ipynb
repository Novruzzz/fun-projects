```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, Ridge
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
boston_df = pd.read_csv('boston.csv')
```

```python
X = boston_df.drop('MEDV', axis=1).to_numpy()
y = boston_df['MEDV'].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
pipe = Pipeline([
    ('polyF', PolynomialFeatures(include_bias=False)),
    ('scaler', StandardScaler()),
    ('lassoLR', Lasso())])
```

```python
params = {'polyF__degree': [1,2,3,4],
          'lassoLR__alpha': np.arange(0.001, 0.011, 0.001)}
```

```python
my_grid = GridSearchCV(pipe, params, scoring='neg_mean_squared_error', verbose=3)
```

```python
my_grid.fit(X_train, y_train)
```

```python
print(my_grid.best_params_)
```

```python
print(-my_grid.best_score_)
```

```python
pipe = Pipeline([
    ('polyF', PolynomialFeatures(include_bias=False)),
    ('scaler', StandardScaler()),
    ('ridgeLR', Ridge())])
```

```python
params = {'polyF__degree': [1,2,3,4],
          'ridgeLR__alpha': np.arange(0.5, 0.6, 0.01)}
```

```python
my_grid = GridSearchCV(pipe, params, scoring='neg_mean_squared_error', verbose=3)
```

```python
my_grid.fit(X_train, y_train)
```

```python
print(my_grid.best_params_)
```

```python
print(-my_grid.best_score_)
```
