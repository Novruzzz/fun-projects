```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
x = np.arange(0, 200)/200
```

```python
noise = np.random.uniform(-0.03, 0.03, (200,))
```

```python
y = 0.4*x + 0.7 + noise
```

```python
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
plt.scatter(x_train, y_train, marker='o', color='g')
plt.scatter(x_val, y_val, marker='o', color='b')
```

```python
my_model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1, activation='linear')
])
```

```python
my_model.summary()
```

```python
y_pred_val = my_model.predict(x_val)
```

```python
plt.figure()
plt.scatter(x_train, y_train, marker='o', color='g')
plt.scatter(x_val, y_val, marker='o', color='b')
plt.plot(x_val, y_pred_val, color='r')
```

```python
my_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.04),
                 loss="mean_squared_error")
```

```python
history = my_model.fit(x=x_train,
                        y=y_train,
                        batch_size=200,
                        epochs=200,
                        verbose=1,
                       validation_data = (x_val, y_val))
```

```python
print(history.history)
```

```python
plt.plot(history.history['loss'], color='g')
plt.plot(history.history['val_loss'], color='b')
```
