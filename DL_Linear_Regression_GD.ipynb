```python
alpha = 0.05
for i in range(500):
    #forward pass
    y_model = w*x+b
    #loss vlaue calculation
    loss_value = np.average((y-y_model)**2)
    print(loss_value)
    #gradients of w and b for the gradient descent
    grad_w = -(2/x.shape[0])*np.sum(x*(y-y_model))
    grad_b = -(2/x.shape[0])*np.sum((y-y_model))
    #updating w and b using their gradients and alpha
    w = w - alpha*grad_w
    b = b - alpha*grad_b
```

```python
noise = np.random.uniform(0, 0.07, (100,))
```

```python
y = 0.4*x + 0.7 + noise
```

```python
w = np.random.random()
b = np.random.random()
```

```python
print('w', w)
print('b', b)
```

```python
def plot_func():
    plt.scatter(x, y, marker='o')
    plt.plot(x, y_model, color='r')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('linear regression using GD')
    plt.show()
    plt.pause(0.1)
```

```python
plot_func()
```

```python
def train_model(x, y, w, b, alpha=0.05, nb_iter=500):
    all_losses = []
    for i in range(nb_iter):
        #forward pass
        y_model = w*x+b
        #loss vlaue calculation
        loss_value = np.average((y-y_model)**2)
        all_losses.append(loss_value)
        #gradients of w and b for the gradient descent
        grad_w = -(2/x.shape[0])*np.sum(x*(y-y_model))
        grad_b = -(2/x.shape[0])*np.sum((y-y_model))
        #updating w and b using their gradients and alpha
        w = w - alpha*grad_w
        b = b - alpha*grad_b

        if i%10 == 0:
            plot_func()
```

```python
train_model(x, y, w, b, alpha=0.05, nb_iter=500)
```
